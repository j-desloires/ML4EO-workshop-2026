```python
# Project root (relative to demo directory where this notebook runs)
project_root = Path("..")

# TODO: Update these paths for the zonal classification task
# Config file for the multi-class zonal model
config_file = project_root / "configs/models/<******_zonal>.yaml"  # Multi-class zonal model config

# Data paths
data_root = project_root / "data/<******>"
test_images = data_root / "test/images"  # Same RGB images as notebook 1.1
test_labels = data_root / "test/<******_zonal_labels>"  # NEW: Different labels file with 4 classes
test_dtm = data_root / "test/dtm"  # Same DEM data as notebook 1.1

# Pre-trained checkpoint for zonal classification
checkpoint_path = project_root / "data/<******_zonal>.ckpt"  # Multi-class zonal model checkpoint

# Output directories
output_dir = Path("./output_zonal")
inference_output = output_dir / "inference_results"
os.makedirs(inference_output, exist_ok=True)

print(f"Config file: {config_file}")
print(f"Checkpoint: {checkpoint_path}")
print(f"Test images: {test_images}")
print(f"Test labels: {test_labels}")
print(f"Test DEM: {test_dtm}")
print(f"Output directory: {inference_output}")
print("\n✓ Setup complete!")
```

<div style="font-size: 18px; line-height: 1.8; margin-bottom: 20px;">
Now we'll run tiled inference on the test dataset using the multi-class zonal classification model. Tiled inference processes large images in smaller chunks, which is essential for handling high-resolution geospatial data.

<div style="font-size: 24px; font-weight: 600; margin: 30px 0 20px 0; line-height: 1.5; color: #3c3c3c;">
3.1 Prepare Inference Command
</div>

<div style="font-size: 18px; line-height: 1.8; margin-bottom: 20px;">
<strong style="color: #0066cc; font-weight: 600;">TODO:</strong> Construct the prediction command using the terratorch CLI. Make sure to:
</div>

<ul style="font-size: 18px; line-height: 1.8; margin: 15px 0 20px 30px;">
<li style="margin-bottom: 12px;">Use the <code style="font-size: 16px; padding: 3px 6px; background-color: #f5f5f5; border-radius: 4px;">terratorch predict</code> command</li>
<li style="margin-bottom: 12px;">Specify the config file and checkpoint path</li>
<li style="margin-bottom: 12px;">Override the data paths for RGB and DEM inputs</li>
<li style="margin-bottom: 12px;">Set the output directory for predictions</li>
</ul>

```python
# TODO: Construct the prediction command
predict_command = f"""terratorch predict \
    --config {config_file} \
    --ckpt_path {checkpoint_path} \
    --trainer.default_root_dir {output_dir} \
    --predict_output_dir {inference_output} \
    --data.init_args.predict_data_root.RGB {test_images} \
    --data.init_args.predict_data_root.DEM {test_dtm}"""

print("Prediction command:")
print(predict_command)
```

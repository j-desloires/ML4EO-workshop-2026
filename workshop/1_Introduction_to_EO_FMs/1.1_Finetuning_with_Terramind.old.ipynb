<div style="font-size: 32px; font-weight: 600; margin: 35px 0 25px 0; line-height: 1.4; color: #2c2c2c;">
1️⃣ Introduction to Fine-Tuning
</div>

<div style="font-size: 18px; line-height: 1.8; margin-bottom: 20px;">
Fine-tuning is the process of adapting a pre-trained foundation model to a specific task or domain. This approach leverages the knowledge learned during pre-training while specializing the model for your use case.
</div>

<div style="font-size: 24px; font-weight: 600; margin: 30px 0 20px 0; line-height: 1.5; color: #3c3c3c;">
1.1. Why Fine-Tune?
</div>

<ul style="font-size: 18px; line-height: 1.8; margin: 15px 0 20px 30px;">
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">Transfer Learning</strong>: Leverage pre-trained knowledge from large datasets</li>
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">Efficiency</strong>: Requires less data and computational resources than training from scratch</li>
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">Performance</strong>: Often achieves better results on specific tasks</li>
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">Adaptability</strong>: Customize models for domain-specific applications</li>
</ul>

<div style="font-size: 24px; font-weight: 600; margin: 35px 0 25px 0; line-height: 1.4; color: #2c2c2c;">1.2. Prepare Fine-Tuning Command</div>

<div style="font-size: 18px; line-height: 1.8; margin-bottom: 20px;">
We'll run terratorch fit for 1 epoch on the smallest available dataset. The config file specifies all the training parameters.

Construct the fine-tuning command with data path overrides. We need to override the data paths since the config has relative paths from project root.

For demo purposes, we add parameters to speed up training:
</div>
<ul style="font-size: 18px; line-height: 1.8; margin: 15px 0 20px 30px;">
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">limit_train_batches</strong>: Only use a fraction of training data per epoch</li>
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">limit_val_batches</strong>: Only use a fraction of validation data</li>
<li style="margin-bottom: 12px;"><strong style="color: #0066cc; font-weight: 600;">fast_dev_run</strong>: Quick sanity check (optional, comment out for actual training)</li>
</ul>

<div style="font-size: 18px; line-height: 1.8; margin-bottom: 20px;">
We are using Terratorch's built-in command line interface to run the each Terratorch command. To do this, we prepare a command string, print it out to validate it is as expected, then excute it in the following cell.
</div>

```python
fine_tuning_command = f"""terratorch fit \
    --config {config_file} \
    --trainer.max_epochs 1 \
    --trainer.default_root_dir {output_dir} \
    --trainer.limit_train_batches 10 \
    --trainer.limit_val_batches 5 \
    --data.init_args.batch_size 4 \
    --data.init_args.train_data_root.RGB {train_images} \
    --data.init_args.train_data_root.DEM {train_dtm} \
    --data.init_args.train_label_data_root {train_labels} \
    --data.init_args.val_data_root.RGB {val_images} \
    --data.init_args.val_data_root.DEM {val_dtm} \
    --data.init_args.val_label_data_root {val_labels}"""

print("Fine-tuning command:")
print(fine_tuning_command)
print("\nNote: Using max_epochs=1, limit_train_batches=10 and limit_val_batches=5 for quick demo.")
print("Remove these limits for full training.")
```

### 1.2 Run Fine-Tuning

Execute the fine-tuning command. This will train the model for 1 epoch and save checkpoints to `./output/lightning_logs/version_0/checkpoints/`.

**Note:** PyTorch Lightning automatically creates the directory structure:
- `lightning_logs/` - Lightning's log directory
- `version_0/` - First run (increments for subsequent runs)
- `checkpoints/` - Contains the `.ckpt` files

```python
# Run fine-tuning (expected execution time: 1m 29.1s)
!{fine_tuning_command}
```

### 1.3 Locate the New Checkpoint

After training, let's find the checkpoint that was just created.

```python
# Find the newly created checkpoint
new_checkpoint_dir = output_dir / "lightning_logs/version_0/checkpoints"
if new_checkpoint_dir.exists():
    new_checkpoints = list(new_checkpoint_dir.glob("*.ckpt"))
    if new_checkpoints:
        print(f"Found {len(new_checkpoints)} checkpoint(s):")
        for ckpt in new_checkpoints:
            print(f"  - {ckpt.name}")
        # Use the first checkpoint for subsequent steps if needed
        new_checkpoint_path = new_checkpoints[0]
    else:
        print("No checkpoints found in the directory.")
else:
    print(f"Checkpoint directory not found: {new_checkpoint_dir}")
```

## 2. Testing

### 2.1 Run Test on Pre-trained Checkpoint

Now we'll test the model using the pre-trained checkpoint on the test dataset.

Construct the test command with data path overrides. We need to override the data paths since the config has relative paths from project root

```python

test_command = f"terratorch test --config {config_file}\
    --ckpt_path {checkpoint_path} \
    --trainer.default_root_dir {output_dir} \
    --data.init_args.test_data_root.RGB {test_images} \
    --data.init_args.test_data_root.DEM {test_dtm} \
    --data.init_args.test_label_data_root {test_labels}"
    
print("Test command:")
print(test_command)
```

```python
# Run testing (expected execution time: 31.2s)
!{test_command}
```

## 3. Prediction

### 3.1 Run Inference

Run prediction on the test images using the pre-trained checkpoint. Let's construct the prediction command.

```python

predict_command = f"terratorch predict \
    --config {config_file} \
    --ckpt_path {checkpoint_path} --trainer.default_root_dir {output_dir} \
    --predict_output_dir {inference_output} \
    --data.init_args.predict_data_root.RGB {test_images} \
    --data.init_args.predict_data_root.DEM {test_dtm}"

print("Prediction command:")
print(predict_command)
```

```python
# Run prediction (expected execution time: 29.9s)
!{predict_command}
```
